```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
```

```python
Attrition = pd.read_csv('general_data.csv')
Attrition.head()
```

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
Attrition.Department = enc.fit_transform(Attrition.Department)
```

```python
y = Attrition.Attrition
X = Attrition.loc[:,['Department','TrainingTimesLastYear','YearsSinceLastPromotion']]
```

```python
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X,y, stratify=Attrition.Attrition,random_state=20) 
model=DecisionTreeClassifier(random_state=10, max_depth=5)
###Decision trees in scikit-learn are implemented in the DecisionTreeRegressor 
##and DecisionTreeClassifier classes. Scikit-learn only implements pre-pruning, not post- pruning.
model.fit(X_train,y_train)
print("accuracy on training set: %f" % model.score(X_train, y_train))
print('\n'"accuracy on test set: %f" % model.score(X_test, y_test))
y_predict = model.predict(X_test)
pd.crosstab(y_test,y_predict)
```
