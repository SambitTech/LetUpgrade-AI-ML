```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
bank = pd.read_excel('Bank_Personal_Loan_Modelling.xlsx')
bank.head()
```

```python
bank = bank.rename(columns = {'Personal Loan':'Personal_Loan','Securities Account':'Securities_Account','CD Account':'CD_Account'})
bank.head()
```

```python
y = bank.Personal_Loan
X = bank.loc[:,['Income','CCAvg','CD_Account']]
```

```python
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X,y, stratify=bank.Personal_Loan,random_state=20) 
model=DecisionTreeClassifier(random_state=10, max_depth=5)
###Decision trees in scikit-learn are implemented in the DecisionTreeRegressor 
##and DecisionTreeClassifier classes. Scikit-learn only implements pre-pruning, not post- pruning.
model.fit(X_train,y_train)
print("accuracy on training set: %f" % model.score(X_train, y_train))
print('\n'"accuracy on test set: %f" % model.score(X_test, y_test))
y_predict = model.predict(X_test)
pd.crosstab(y_test,y_predict)
```
